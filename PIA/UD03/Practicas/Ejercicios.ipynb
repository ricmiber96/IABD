### Ejercicio 1
Crea un programa Elecciones.py que va a representar en un gráfico los
resultados de unas elecciones a las que concurren cuatro partidos políticos, PA,
PB, PC y PD. El usuario deberá introducir en una línea los votos recibidos, en
forma de números enteros que representan a cada partido (1 = PA, 2 = PB, 3 =
PC, 4 = PD, por ejemplo: 1 1 2 1 1 2 4 3 2 1), y mostrará un gráfico circular con
el resultado final de las elecciones. Junto al nombre de cada partido debe
aparecer el porcentaje de voto total. Cualquier número que no esté en el rango
indicado (de 1 a 4) deberá descartarse del conteo final.

```python
import matplotlib.pyplot as plt
```

```python
votos_user = input("Introduce los votos (separados por espacios): ")
num_votos = [0, 0, 0, 0]
array_partidos = ["PA", "PB", "PC", "PD"] 

for n in votos_user:
    if n == ' ':
     continue
    if int(n) == 1:
       num_votos[0] += 1
    if int(n) == 2:
       num_votos[1] += 1
    if int(n) == 3:
       num_votos[2] += 1
    if int(n) == 4:
       num_votos[3] += 1

label_pie_percentage = []

for i in range(len(num_votos)):
    label_pie_percentage.append(array_partidos[i] + " (" + str((num_votos[i]*100)/sum(num_votos)) + "%)")


plt.pie(num_votos, labels=label_pie_percentage)
plt.show()



```

## Ejercicio 2
Crea un programa Ventas.py que le pregunte al usuario por las ventas de una
empresa en un rango de años (indicado por el usuario) y muestre un diagrama
de líneas con la evolución.

```python
array_years = []
array_ventas = []

input_rango_year = input("Introduce el rango de años yyyy / yyyy: ").split("/")
years_array = input_rango_year

for i in range(int(years_array[0]), int(years_array[1])+1):
    input_venta_year = input(f"Introduce ventas para el año {str(i)}")
    array_years.append(i)
    array_ventas.append(int(input_venta_year))


plt.bar(array_years, array_ventas, align="center", color="skyblue")
plt.xticks(array_years, labels=array_years)
plt.title("Grafica de ventas anuales")
plt.xlabel("Años")
plt.ylabel("Ventas")
plt.show()

```

## Ejercicio 3
Crea un programa GraficoCircularNotas.py que le pida al usuario una
secuencia de notas numéricas (en una línea, separadas por espacios) y muestre
un gráfico circular con el porcentaje de suspensos, aprobados, notables y
sobresalientes. Personaliza el color de cada porción del gráfico a tu gusto. Por
ejemplo:


```python
notas_array = input("Introduce las notas (separadas por espacios): ").split(" ")
num_notas = [0, 0, 0, 0]
labels_notas = ['Suspensos', 'Aprobados', 'Notables', 'Sobresalientes']

for nota in notas_array:
    if nota == ' ':
        continue
    if int(nota) < 5:
        num_notas[0] += 1
    elif int(nota) >= 5 and int(nota) <= 6:
        num_notas[1] += 1
    elif int(nota) >= 7 and int(nota) <= 8:
        num_notas[2] += 1
    elif int(nota) >= 9 and int(nota) <= 10:
        num_notas[3] += 1

array_labels_notas = []
for i in range(len(num_notas)):
    array_labels_notas.append(labels_notas[i] + " (" + str(num_notas[i]) + ")")

colors_notas = ['red', 'yellow', 'green', 'blue']
plt.pie(num_notas, labels=array_labels_notas,colors=colors_notas, startangle=90)
plt.title("Distribución de Notas")
plt.show()


    
```

## Ejercicio 4
Crea un programa llamado GraficoNotas.py que le pida al usuario una
secuencia de notas numéricas de dos alumnos (el mismo número de notas para
ambos), separadas por espacios, junto con sus nombres (en otra línea aparte).
Después muestra en un gráfico de
líneas las notas introducidas. Utiliza
colores y marcadores diferentes para
cada alumno, y una leyenda con sus
nombres. Debe quedarte algo así:

```python
notas_alumno1 = list(map(float,input("Introduce las notas del alumno 1 (separadas por espacios): ").split(" ")))
notas_alumno2 = list(map(float,input("Introduce las notas del alumno 2 (separadas por espacios): ").split(" ")))
nombres_alumnos = input("Introduce los nombres de los alumnos (separados por espacios): ").split(" ")

if len(notas_alumno1) != len(notas_alumno2):
    print("Error: Los alumnos deben tener el mismo número de notas.")

else:
    x = list(range(0,len(notas_alumno1)))
    nombres_alumnos = nombres_alumnos

    plt.plot(x, notas_alumno1, color="blue", marker="o")
    plt.plot(x, notas_alumno2, color="red", marker="x")

    plt.ylim(0, 10)
    plt.xticks(range(0, len(x)))

    plt.ylabel("Notas")
    plt.xlabel("Asignaturas")
    plt.legend(nombres_alumnos, loc="lower center")
    plt.title("Notas de los alumnos")
    plt.show()

```

## Ejercicio 5
Crea un programa MultiplesGraficos.py con un panel 2x2 para mostrar 4
gráficos diferentes sobre el dataset parques_eolicos.csv. En concreto se piden los
siguientes gráficos:
- Gráfico circular que muestre la proporción de parques eólicos por
provincia
- Gráfico de barras con la potencia media por provincia
- Histograma con la frecuencia del número de aerogeneradores
- Gráfico de cajas que muestre la dispersión del número de
aerogeneradores en Valladolid y Palencia

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('datasets\parques_eolicos.csv', delimiter=';')
print(data.head())

provincias_df = data['provincia'].value_counts()
provincias_df = provincias_df.reset_index()
provincias_df.columns = ['Provincia', 'Numero_parques']
provincias_df_potencia_med = data.groupby('provincia')['potencia'].mean()

num_parques = provincias_df['Numero_parques']
num_provincias = len(provincias_df)
df_provincia_VA = data[data['provincia'] == 'VA']['aerogeneradores']
df_provincia_PA = data[data['provincia'] == 'PA']['aerogeneradores']

fig, axes = plt.subplots(nrows= 2, ncols=2, figsize=(10, 8))
axes[0][0].pie(provincias_df['Numero_parques'], labels=provincias_df['Provincia'])
axes[0][0].set_title('Parques por provincia')
axes[0][1].bar(provincias_df['Provincia'], provincias_df_potencia_med, color='blue')
axes[0][1].set_title('Potencia media por provincia')
axes[1][0].hist(data['aerogeneradores'], bins=8, color='blue', edgecolor='black', alpha=0.7)
axes[1][0].set_title('Numero de aerogeneradores')
axes[1][1].boxplot([df_provincia_VA,df_provincia_PA], labels=['VA','PA'])
axes[1][1].set_title('Aerogeneradores en VA y PA')
plt.show()

```
