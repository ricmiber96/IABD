```python
import matplotlib.pyplot as plt
#Ejemplo de ejecuci√≥n
alumnado = ['Alumno 1', 'Alumno 2', 'Alumno 3', 'Alumno 4' ]
mate = [1,2,7,9]
lengua = [3,5,8,9]
```

```python
plt.figure(figsize=(4,2))
plt.plot(alumnado, mate)
plt.plot(alumnado, lengua)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
axes[0].plot(alumnado, mate, color='red')
axes[1].plot(alumnado, lengua, color='darkblue')
plt.show()
```
