 - Los **hiperparámetros** son **valores configurables** que no se aprenden directamente durante el entrenamiento, sino que se establecen antes de iniciar el proceso de entrenamiento. Estos parámetros **controlan cómo se entrena el modelo** y afectan al rendimiento del modelo.

En el **método wrapper hacia adelante** (forward selection), comenzamos con un conjunto vacío de características y, en cada iteración, añadimos una característica que maximiza el rendimiento del modelo entre las características restantes. 3.  **Paso 3**: Ahora tenemos 2 características seleccionadas y 8 restantes. Probamos cada una de las 8 características restantes junto con las 2 seleccionadas, lo cual requiere **8 modelos**.

**a) ¿Porqué no es buena idea utilizar todos los datos para entrenar al modelo? Si lo hacemos y su comportamiento con estos datos es bueno, por ejemplo acierta el 96% de las veces ¿No es suficiente para decir que es un buen modelo? ¿O hay algo que no sabemos aún para darlo por bueno?** 
Usar todos los datos para entrenar un modelo puede llevar a una problemática llamada sobreajuste (overfitting). Cuando el modelo se entrena y ajusta únicamente sobre un conjunto de datos específico, puede llegar a "memorizar" estos datos en lugar de aprender patrones generales. Como resultado, el modelo puede presentar un desempeño excelente en los datos de entrenamiento (por ejemplo, una precisión del 96%), pero este rendimiento no necesariamente se traduce a datos que no ha visto antes, que es el objetivo final.

**b) ¿Porqué nos puede interesar hacer 3 divisiones de los datos (train+validación+test) en ves de dos (train+test)?** 
Si solo se usan datos de entrenamiento y prueba, el ajuste de hiperparámetros y la optimización del modelo se realizarían sobre los datos de prueba, lo cual podría introducir un sesgo hacia estos datos, afectando la objetividad de la evaluación final.

**c) ¿Qué porcentaje se suelen usar para cada partición de datos?** 
En una división típica se suele usar un 70% de los datos para entrenamiento, un 15% para validación y un 15% para prueba. Esto busca asegurar suficientes datos para cada tarea y optimizar tanto el aprendizaje como la evaluación.

**d) Si no tenemos muchos datos, en vez de hacer la partición en 3 datasets (train+validación+prueba) ¿Qué otros métodos se pueden utilizar?**
Cuando se dispone de pocos datos, algunas técnicas alternativas:

 - **Validación cruzada (cross-validation)**: en lugar de hacer una división fija, se realizan múltiples divisiones de los datos y se entrena y evalúa el modelo en cada división.
 - **Leave-One-Out Cross-Validation (LOOCV)**: es una variación de la validación cruzada en la que en cada iteración se usa solo un ejemplo como conjunto de prueba, y el resto como conjunto de entrenamiento. Este método es útil con muy pocos datos, pero puede ser computacionalmente costoso.
 - **Aumento de datos (data augmentation)**: en problemas como el procesamiento de imágenes o texto, es posible generar variaciones del mismo dato (por ejemplo, rotando o escalando imágenes) para incrementar el tamaño del conjunto de datos y hacer el modelo más robusto.
